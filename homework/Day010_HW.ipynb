## requests + BeautifulSoup


```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.dcard.tw/f'

k = requests.get(url)
response = k.text
r = BeautifulSoup(response)
print(response)

r.encoding = 'utf-8'
print(r.text[0:3000])
```

## Grab + PyQuery

```python
from grab import Grab
g = Grab()
resp = g.go('https://www.dcard.tw/f')
resp.body

from pyquery import PyQuery as pq
doc = pq(resp.body)
title = doc('title')
print(type(title), title.text())
```
