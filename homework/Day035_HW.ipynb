## 作業目標

* 找一個之前實作過的爬蟲改用多線程改寫，比較前後時間的差異。





```python
import requests
import time
from bs4 import BeautifulSoup

urls = ['https://www.ettoday.net/news/news-list.htm','https://www.ettoday.net/news/news-list.htm']

startTime = time.time()

for url in urls:
    r = requests.get(url)
    soup = BeautifulSoup(r.text, "html5lib")

    for d in soup.find(class_="part_list_2").find_all('h3'):
        print(d.find(class_='date').text, d.find_all('a')[-1].text)
    
finishTime = time.time()
print(finishTime - startTime) # 正常情況的爬蟲所需時間
```

```python
import requests
import time
import _thread
from bs4 import BeautifulSoup

urls = ['https://www.ettoday.net/news/news-list.htm','https://www.ettoday.net/news/news-list.htm']

startTime = time.time()
    
for url in urls:
    _thread.start_new_thread( requests.get, (url, ) )
    r = requests.get(url)
    soup = BeautifulSoup(r.text, "html5lib")

    for d in soup.find(class_="part_list_2").find_all('h3'):
        print(d.find(class_='date').text, d.find_all('a')[-1].text)
                             
finishTime = time.time()
print(finishTime - startTime) # 多線程爬蟲所需時間
```
