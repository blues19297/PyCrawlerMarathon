## 作業目標

根據範例：

1. 取出今天所有的新聞
2. 取出現在時間兩小時內的新聞

## 1. 取出今天所有的新聞

```python
from selenium import webdriver
from bs4 import BeautifulSoup

browser = webdriver.Chrome(executable_path = 'chromedriver/chromedriver')
browser.get("https://www.ettoday.net/news/news-list.htm")

import time
import datetime

for i in range(58):
    time.sleep(1)
    browser.execute_script("window.scrollTo(0, 100000);")
    
html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    if "2020/07/29" in d.find(class_="date").text:
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

## 2. 取出現在時間兩小時內的新聞

```python
from datetime import datetime,timedelta
for d in soup.find(class_="part_list_2").find_all('h3'):
    k = datetime.strptime(d.find(class_="date").text,'%Y/%m/%d %H:%M')
    if k >= (datetime.now()-timedelta(hours=2)):
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
browser = webdriver.Chrome(executable_path = 'chromedriver/chromedriver')
browser.get("https://www.ettoday.net/news/news-list-2020-7-26-0.htm")

for i in range(36):
    time.sleep(1)
    browser.execute_script("window.scrollTo(0, 50000);")
    
html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")

three_days_ago_time = datetime.now()-timedelta(days=3)
three_days_ago_month = three_days_ago_time.strftime("%m").replace("0","")
three_days_ago_day = three_days_ago_time.strftime("%d").replace("0","")
three_days_ago_15_clock = datetime.strptime(three_days_ago_time.strftime("%Y/%m/%d") + ' 15', '%Y/%m/%d %H')
three_days_ago_17_clock = datetime.strptime(three_days_ago_time.strftime("%Y/%m/%d") + ' 17', '%Y/%m/%d %H')

for d in soup.find(class_="part_list_2").find_all('h3'):
    k = datetime.strptime(d.find(class_="date").text,'%Y/%m/%d %H:%M')
    if three_days_ago_15_clock <= k <= three_days_ago_17_clock:
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
```
