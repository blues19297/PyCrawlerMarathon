### 1. 根據範例使用 selenuim 取出 Hahow 網站中所有課程的資料，並用一個適合的資料結構儲存（Hint: 將所有資料存在一個變數當中取出）

```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from bs4 import BeautifulSoup

browser = webdriver.Chrome(executable_path='chromedriver/chromedriver')
browser.get("https://hahow.in/courses")
```

```python
# 取得資料（等瀏覽器資料出現後才執行）
html_source = browser.page_source
soup = BeautifulSoup(html_source, 'html.parser')
```

```python
course = []

for title in soup.find_all("h4",class_="title"):
    d = list(title)
    course += d
    
print(course)
```

```python
browser.close()
```

### 2. 剛剛的做法只能取出第一頁，試著取出所有頁面的資料。

```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from bs4 import BeautifulSoup

browser = webdriver.Chrome(executable_path='chromedriver/chromedriver')
browser.get("https://hahow.in/courses")
```

```python
# 取得資料（等瀏覽器資料出現後才執行）
html_source = browser.page_source
soup = BeautifulSoup(html_source, 'html.parser')
```

```python
# 1. 先取出全部有幾頁

count = 0

for i in range(1,22):
    url = "https://hahow.in/courses?page=" + str(i)
    browser.get(url)
    count += 1
        
print(count)
```

```python
# 2. 每頁都取出資料放入 courses
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from bs4 import BeautifulSoup
courses = []
browser = webdriver.Chrome(executable_path='chromedriver/chromedriver')

for i in range(1,22):
    url = "https://hahow.in/courses?page=" + str(i)
    browser.get(url)
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, 'html.parser')
    for title in soup.find_all("h4",class_="title"):
        d = list(title)
        courses += d
    
browser.close()
```

```python
# 3. done!

print(courses)
```
